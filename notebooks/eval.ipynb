```python
from src.evaluation import score_answer
import pickle as pkl
with open('../results/full_run_new.pkl', 'rb') as f:
    data = pkl.load(f)

import pandas as pd
import os
from pathlib import Path

dataset_path = Path("~/mnt/shared-resources-sentient-research/salah_resources/datasets/officeqa/")
data_file = "officeqa.csv"
original_data = pd.read_csv(os.path.join(dataset_path, data_file))
```

```python
eval_set_original = {
    'question': [],
    'index': [],
    'predicted': [],
    'difficulty': [],
    'ground_truth': [],
    'score_0.05': [],
    'score_0.01': [],
    'score_0.1': [],
    'score_0.0': [],
    'score_0.025': [],
    "final_score": []
}
```

```python
for example in data:
    try:
        question = example.question
        index = example.index
        ground_truth = example.ground_truth
        predicted = example.trace.output.final_answer
        difficulty = original_data.iloc[index]['difficulty']
        final_score = 0.0
        for tolerance in [0.05, 0.01, 0.1, 0.0, 0.025]:
            score = score_answer(ground_truth=ground_truth, predicted=predicted, tolerance=tolerance)
            final_score += score
            eval_set_original[f'score_{tolerance}'].append(score)
        final_score /= 5
        eval_set_original['final_score'].append(final_score)
        eval_set_original['question'].append(question)
        eval_set_original['index'].append(index)
        eval_set_original['ground_truth'].append(ground_truth)
        eval_set_original['predicted'].append(predicted)
        eval_set_original['difficulty'].append(difficulty)
    except Exception as e:
        print(f"Error processing example {index}: {e}")
        continue
```

```python
import pandas as pd
original_df = pd.DataFrame(eval_set_original)
```

```python
print(f"0 tolerance score: {original_df['score_0.0'].mean().item()}")
print(f"final score averaged: {original_df['final_score'].mean().item()}")
```

```python
original_df.to_csv("../.dataset/solved_dataset.csv", index=False)
```
